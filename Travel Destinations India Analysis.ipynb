```python
# Drop unnecessary Columns

travel_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
travel_data['Weekly Off'].value_counts()
```

```python
travel_data['Weekly Off'].fillna('ALL day open', inplace=True)
```

```python
travel_data.head()
```

```python
travel_data["State"] = travel_data["State"].replace("Maharastra", "Maharashtra")
```

```python
# Rename confusing column headers.

travel_data.rename(columns={
    'time needed to visit in hrs': 'Time Needed (hrs)',
    'Google review rating': 'Google Rating',
    'Entrance Fee in INR': 'Entrance Fee (INR)',
    'Airport with 50km Radius': 'Nearby Airport',
    'Number of google review in lakhs': 'Google Reviews (Lakhs)',
    'Best Time to visit': 'Best Time'
}, inplace=True)

travel_data.head()
```

```python
# Replace 'Unknown' or missing values in 'Establishment Year' with NaN and convert to numeric.

travel_data['Establishment Year'].replace('Unknown', pd.NA, inplace=True)
travel_data['Establishment Year'] = pd.to_numeric(travel_data['Establishment Year'], errors='coerce')

travel_data.head(20)
# 16th value showed 'unknown' which is changed to 'NaN'.
```

```python
# All places in Gujarat

travel_data[travel_data['State'] == 'Gujarat']
```

```python
# All places in Western Zone

travel_data[travel_data['Zone'] == 'Western']
```

```python
# Highly-rated & popular places.

travel_data[(travel_data['Google Rating'] > 4.5) & (travel_data['Google Reviews (Lakhs)'] > 4)]
```

```python
# Free or very low-cost places (under â‚¹50)

travel_data[travel_data['Entrance Fee (INR)'] <= 50].sample(n = 5)
```

```python
travel_data["Zone"].value_counts()
```

```python
# Distribution of Landmarks by Zone

plt.figure(figsize=(10, 6), dpi=200)
sns.countplot(x='Zone', data=travel_data, order = travel_data['Zone'].value_counts().index)
plt.title('Distribution of Landmarks by Zone')
plt.xlabel('Zone')
plt.ylabel('Number of Landmarks')
plt.show()
```

```python
travel_data["State"].value_counts()
```

```python
# Distribution of Landmarks by State (Top 10)

plt.figure(figsize=(10, 6), dpi=200)
state_counts = travel_data['State'].value_counts()
sns.barplot(x=state_counts, y=state_counts.index, color = "grey")
plt.title('States by Number of Landmarks')
plt.xlabel('Number of Landmarks')
plt.ylabel('State')
plt.show()
```

```python
# Types of Landmarks
plt.figure(figsize=(12, 8), dpi=200)
type_counts = travel_data['Type'].value_counts().head(20)
sns.barplot(x=type_counts, y=type_counts.index, color = "r")
plt.title('Top 20 Types of Landmarks')
plt.xlabel('Number of Landmarks')
plt.ylabel('Type')
plt.show()
```

```python
travel_data["Google Rating"].value_counts()
```

```python
# Review Ratings and Entrance Fees
plt.figure(figsize=(10, 6),dpi=200)
sns.scatterplot(data=travel_data, x='Google Rating', y='Entrance Fee (INR)', color = "g")
plt.title('Review Ratings vs. Entrance Fees')
plt.xlabel('Google Rating')
plt.ylabel('Entrance Fee (INR)')
plt.show()
```

```python
# Time Needed to Visit
plt.figure(figsize=(10, 6), dpi=200)
sns.histplot(travel_data['Time Needed (hrs)'], bins=20, kde=True, color = "y")
plt.title('Distribution of Time Needed to Visit')
plt.xlabel('Time Needed to Visit (hrs)')
plt.ylabel('Number of Landmarks')
plt.show()
```

```python
# Filtering top-rated and popular spots(4.5+ rating, 2L+ reviews)
filtered_df = travel_data[(travel_data['Google Rating'] >= 4.5) & (travel_data['Google Reviews (Lakhs)'] >= 2)]
```

```python
filtered_df
```

```python
#Top-rated and popular spots(4.5+ rating, 2L+ reviews) ensure a safe, memorable visit.
plt.figure(figsize=(10, 6))
sns.barplot(data=filtered_df, y='Name', x='Google Rating', palette='viridis')
plt.title('Top Rated Places with More Than 2 lakh Google Reviews')
plt.xlabel('Google Rating')
plt.ylabel('Place Name')
plt.show()
```

```python
top10_types = travel_data['Type'].value_counts().head(10).index

# Filter the dataframe to only include these top 20 types
filtered_top10_df = travel_data[travel_data['Type'].isin(top10_types)]
```

```python
plt.figure(figsize=(14, 8), dpi=200)
sns.countplot(data=filtered_top10_df,y='Type',hue='Zone',
              order=top10_types,
              palette='Set1')
plt.title('Type of Attractions by Region', fontsize=16)
plt.xlabel('Number of Places', fontsize=14)
plt.ylabel('Type of Attraction', fontsize=14)
plt.legend(title='Region')
plt.show()
```
